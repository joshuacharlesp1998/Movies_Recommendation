```python
#Importing the standard and necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings

```

```python
#Storing the data into a variable
movie_md=pd.read_csv('/content/gdrive/MyDrive/Movies Dataset/movies_metadata.csv')
```

```python
#It describes the data 
movie_md.head()
```

```python
#it shows the avaliable columns in the dataframe
movie_md.columns
```

```python
#Here we are filtering out the whole data based on the "votecount"
movie_md=movie_md[movie_md['vote_count']>=55]
```

```python
#we select certain columns and renaming the dataframe
movie_md=movie_md[['genres','original_title','id','overview']]
```

```python
movie_md
```

```python
#we are storing the title in another column using copy() inbuild function
movie_md['Title']=movie_md['original_title'].copy()
```

```python
movie_md
```

```python
#Using this, we can reset the index of the data frame
movie_md.reset_index(inplace=True,drop=True)
```

```python
movie_md.head()
```

```python
#we are following the same steps for the next data set
credits=pd.read_csv('/content/gdrive/MyDrive/Movies Dataset/credits.csv')
```

```python
credits.columns
```

```python
credits
```

```python
credits=credits[['cast','id']]
```

```python
credits
```

```python
keywords=pd.read_csv('/content/gdrive/MyDrive/Movies Dataset/keywords.csv')
```

```python
keywords.head()
```

```python
movie_md
```

```python
#here we are converting the "id" column in to the numeric
movie_md[movie_md['id'].str.isnumeric()]
```

```python
#now we are again converting it into numeric
movie_md['id'] = movie_md['id'].astype(int)
```

```python
#In these steps, we are mergring the whole data frame into single data frame for easy useage
df=pd.merge(movie_md,keywords,on='id',how='left')
```

```python
df=pd.merge(df,credits,on='id',how='left')
```

```python
#again Reseting the index
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df['genres']
```

```python
#In this step we are going to split the "Generes" column in to their names alone. using the line of code, by calling the handy Lambda function
#Lambda function will iterate through each element and gives the required results.
df['genres']=df['genres'].apply (lambda x: [i['name'] for i in eval(x)])
```

```python
#we again use the Lambda function, to join the unnecessary space avaliable in the data field of the column
df['genres']=df['genres'].apply (lambda x: ' '.join([i.replace(" ","") for i in x]))
```

```python
#the final column looks something like this
df['genres']
```

```python
#we need to follow the same procedure for the "Keywords" and "Cast" column
df
```

```python
df['keywords']=df['keywords'].apply (lambda x: [i['name'] for i in eval(x)])
```

```python
df['keywords'].value_counts
```

```python
df['keywords'] = df['keywords'].apply(lambda x: ' '.join([i.replace(" ",'') for i in x]))
df['keywords']
```

```python
df
```

```python
#Now we are checking whether the data having any null values, the null values will affect our model, so its better to remove them
df.isna().sum()
```

```python
df.size
```

```python
#here we are filling the Null values with "[]", where inplace=True will make changes in the original data frame.
df['keywords'].fillna('[]', inplace=True)
```

```python
df
```

```python
df
```

```python
df['cast']=df['cast'].apply(lambda x: [i['name'] for i in eval(x)])
```

```python
df['cast'].fillna('[]', inplace=True)
```

```python
df['cast']
```

```python
df['cast']=df['cast'].apply(lambda x: " ".join([i.replace(" ","") for i in x]))
```

```python
df['cast']
```

```python
#Atlast we have made our dataframe into a string format
df
```

```python
#we are creating new column, that will include Overview, Original_Title, Generes, Keywords, Cast
df['Tags']=df['overview']+" "+df["original_title"]+" "+df['genres']+" "+df['keywords']+" "+df['cast']
```

```python
df
```

```python
#Again we are changing the dataframe into required columns using this line of code
df=df[['id','Title','Tags']]
```

```python
df
```

```python
df.isnull().sum()
```

```python
#It removes the null values in the data frame
df.dropna()
```
